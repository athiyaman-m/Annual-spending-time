```python
dataset = pd.read_csv('Ecommerce Customers')
```

```python
x=dataset.iloc[:,3:7].values  #independent variable
y=dataset.iloc[:,7:].values   #dependent variable
X = dataset.iloc[:,6:7].values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```
